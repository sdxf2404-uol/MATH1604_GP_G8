## Role of the M3 Module

The M3 module has two main functions:

1. `generate_means_sequence(collated_answers_path)`  
   This function calculates the mean answer value for each question.

2. `visualize_data(collated_answers_path, n)`  
   This function creates visualisations of the answer data. If `n = 1`, it produces a scatter plot of the mean answer values. If `n = 2`, it produces a line plot showing the individual answer sequences together with the mean sequence.

```python
from pathlib import Path
import sys

# Locate the project root whether the notebook is run from the root folder
# or from inside the reviews folder.
root = Path.cwd()

if not (root / "scripts").exists():
    root = root.parent

sys.path.insert(0, str(root / "scripts"))

from data_analysis_M3 import load_collated_data, generate_means_sequence, visualize_data

collated_path = root / "output" / "collated_answers.txt"

print("Using collated file:", collated_path)
print("File exists:", collated_path.exists())
```

## Loading the Collated Data

The file `collated_answers.txt` contains the combined answer data from the respondent files.

For analysis, the selected answer for each question is converted into a numerical value:

- 1, 2, 3, or 4 for the selected option
- 0 for an unanswered question

This gives a structured numerical representation of the quiz responses, which can then be used for statistical analysis and visualisation.

```python
data = load_collated_data(collated_path)

number_of_respondents = len(data)
number_of_questions = len(data[0])

print("Number of respondents:", number_of_respondents)
print("Number of questions:", number_of_questions)
```

## Data Validation

Before calculating statistics, I check that the data have the expected structure.

The dataset should contain multiple respondent sequences, and each respondent should have answers for 100 questions. This check is important because the mean sequence only makes sense if every question position has been parsed consistently.

```python
all_lengths = [len(row) for row in data]

print("Unique sequence lengths:", sorted(set(all_lengths)))

if all(length == 100 for length in all_lengths):
    print("Validation passed: each respondent has 100 answers.")
else:
    print("Validation failed: at least one respondent does not have 100 answers.")
```

## Mean Answer Sequence

The function `generate_means_sequence()` calculates the mean answer value for each question.

Unanswered responses are coded as 0, but they are excluded from the mean calculation. This is important because 0 is not one of the four answer options. Including 0 would reduce the mean and distort the result.

```python
means = generate_means_sequence(collated_path)

print("Number of mean values:", len(means))

if len(means) == 100:
    print("Mean sequence check passed: there is one mean value for each question.")
else:
    print("Mean sequence check failed.")
```

```python
print("Minimum mean value:", min(means))
print("Maximum mean value:", max(means))
print("Overall average of mean values:", sum(means) / len(means))
```

## Scatter Plot

The scatter plot shows the mean answer value for each question from 1 to 100.

This visualisation is useful because it gives a quick overview of how the mean answer values vary across the quiz. If the points are randomly scattered with no clear structure, this would suggest weaker evidence of a deliberate pattern. If the points show repeated shapes, cycles, or systematic movement, this could suggest possible structure.

```python
visualize_data(collated_path, 1)
```

## Line Plot

The line plot shows the individual answer sequences together with the mean sequence.

The individual lines are expected to look dense because the dataset contains many respondent files. However, the mean line gives a clearer summary of the overall behaviour across the 100 questions.

This plot is useful for comparing individual variation with the overall mean pattern.

```python
visualize_data(collated_path, 2)
```

## Interpretation of Results

The scatter plot shows that the mean answer values move across a wide range between approximately 1 and 4. This suggests that the answers are not concentrated around one single option throughout the quiz.

The line plot shows strong variation between individual respondents, which is expected because different files may contain different selected answer patterns. The thicker line makes it easier to see the mean sequence, which helps summarise the overall pattern more clearly.

A visual inspection shows that the answer values fluctuate between lower and higher levels across the 100 questions. This suggests that the responses are not evenly distributed or concentrated on a single option. However, visual evidence is insufficient to confirm the presence of any deliberate hidden pattern.

A stronger conclusion would require further analysis, such as checking answer-option frequencies, repeated blocks, cycles, or autocorrelation.

## Critical Reflection

The mean answer value is useful because it provides a simple numerical summary for each question. It allows the group to visualise the answer sequence across all 100 questions.

However, the mean also has limitations. Different distributions of answers can produce similar mean values. For example, a question with many 1s and 4s could have a similar mean to a question with many 2s and 3s, even though the underlying distributions are different.

For this reason, the scatter plot and line plot should be treated as an initial stage of analysis rather than final proof of a pattern. The visualisations are helpful for identifying possible structure, but they should be supported by additional statistical evidence before making a strong claim.
